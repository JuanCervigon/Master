# Introducci칩n

En este notebook se desarrollan cuatro formas formas de estimar los par치metros de una regresi칩n lineal simple ($\beta_0$,$\beta_1$ y $\sigma$). Vamos a utilizar los siguientes m칠todos para calcular los par치metros

* F칩rmulas expl칤citas
* Proyecci칩n ortogonal (matrices)
* Librer칤a Statsmodels de Python
* Librer칤a Scikit-Learn de Python

En esta seci칩n se importan las librer칤as y se genera un conjunto de datos $(X,y)$ sobre los que se va a realizar la estimaci칩n de los par치metros. Primero se genera la variable independiente $x$ y se construye la matriz

\begin{align}
        X = \begin{pmatrix}
        1 & x_1 &  \\
        1 & x_2 & \\
        ... & ... \\
        1 & x_n
        \end{pmatrix}
    \end{align}


Despu칠s, a partir de los par치metros $\beta_0$, $\beta_1$ y $\sigma$ generamos la variable dependiente $y$ con el modelo

\begin{align}
      y_i=\beta_0 + \beta_1 x_i + \epsilon_i
      \end{align}

Donde $\epsilon_i$ son $iid$  $N(0,\sigma^2)$

Despu칠s guardamos el resultado de la simulaci칩n en el vector $y$

\begin{align}
        y = \begin{pmatrix}
        y_1 \\
        y_2 \\
        ...\\
        y_n
        \end{pmatrix}
    \end{align}




```python
# Importar librer칤as
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generar los datos

# Numero de observaciones
observaciones=500

# Valores de x
x=np.random.random(observaciones)*100

# Par치metros del modelo, beta0, beta1 y sigma
parametros=np.array([1,4])
sigma=40

# Variable dependiente
y=parametros[0]+parametros[1]*x + np.random.normal(0,sigma,observaciones)

# Hacer un gr치fico de los datos
plt.scatter(x,y);

```

# F칩rmulas explicitas

Las f칩rmulas utilizadas para la estimaci칩n de los par치metros son:


Primero realizamos los c치lculos intermedios

\begin{align}
      S_{xy}=\frac{1}{n}\sum_{i=1}^{n}x_iy_i-\overline{x}\overline{y}
      \end{align}

\begin{align}
  S_{xx}=\frac{1}{n}\sum_{i=1}^{n}x_i^2-\overline{x}^2
     \end{align}

Para calcular las betas:

\begin{align}
  \beta_1=\frac{S_{xy}}{S_{xx}}
     \end{align}
\begin{align}
  \beta_0=\overline{y}-\beta_1\overline{x}
       \end{align}

El par치metro $\sigma$ lo estimamos como la raiz cuadrada de la suma de los residuos al cuadrado dividido por los grados de libertad



\begin{align}
  \sigma=\sqrt{\frac{\sum_{i=1}^{n}(y_i-\beta_0-\beta_1x_i)^2}{n-2}}
       \end{align}



```python
# F칩rmulas expl칤citas

# Estimaci칩n de beta1
sxy=1/observaciones*np.dot(x,y)-np.mean(x)*np.mean(y)
sxx=1/observaciones*np.dot(x,x)-np.mean(x)**2
beta1=sxy/sxx

# Estimaci칩n de beta0
beta0=np.mean(y)-beta1*np.mean(x)

# Estimaci칩n de sigma^2
residuos=y-beta0-beta1*x
sigma=np.sqrt(sum(residuos**2)/(observaciones-2))

# Guardar resultado
parametros_formulas=[beta0,beta1,sigma]
parametros_formulas
```

# Proyecci칩n ortogonal (matrices)

En este caso proyectamos ortogonalmente el vector de observaciones $y$ en el espacio vectorial generado por el vector $x$ y un vector de $unos$. El vector $y$ que obtenemos de la proyecci칩n ortogonal es $$\hat{y}$$ y los coeficientes de la combinaci칩n lineal de la proyecci칩n son las betas

En f칩rmulas: $$\hat{y}=X\beta$$


PAra ello vamos a construir la matriz $X$ de n filas y dos columnas formada por el vector de unos y la variable independiente $x$, $$X=[unos,x]$$

Entonces el modelo lo podemos escribir as칤

\begin{align}
  y=X\beta+洧띉
   \end{align}

Asumiendo que se verifican las condiciones del modelo regresi칩n lineal (probabil칤stico), entonces los errores son independientes de las observaciones, por lo que se cumplen las condiciones de ortogonalidad con los vectores de la matriz $X$, es decir, $X'洧=0$, (las condiciones de ortogonalidad o de independencia de los residuos son (i) la suma de residuos es igual a cero y (ii) la suma de $x_i洧=0$ residuos son ortogonales a la variable independiente)

Despejando ya f칩rmula anterior teniendo en cuenta que $X'洧=0$, se obtiene

\begin{align}
    \beta=(X'X)^{-1}X'Y
      \end{align}

Lo que nos permite obtener las betas directamente.


Para calcular el par치metro $\sigma$ utilzamos la misma f칩rmula que en el apartado anterior, es decir, la raiz cuadrada de la suma de los residuos al cuadrado dividido por los grados de libertad, como en el apartado anterior


```python
# A침adir un vector de unos para crear la matriz X

# Crear el vector de unos de longitud igual al n칰mero de observaciones
unos=np.ones(len(x))

# Juntar el vector de unos con el vector de observaciones para crear la matriz X
X=np.column_stack((unos,x))

# Obtener las tres primeras filas de la matriz X
X[:3,:]
```

```python
# C치lculo de las betas
betas=np.matmul(np.matmul(np.linalg.inv(np.matmul(X.T,X)),X.T),y)

# C치lculo de sigma
residuos=y-np.matmul(X,betas)
sigma=np.sqrt(sum(residuos**2)/(observaciones-2))

# Guardar resultado
parametros_matrices=np.append([betas], [sigma])
parametros_matrices
```

# Statsmodels

Los par치metros se estiman directamente con la sentencia correspondiente de StatsModels


```python
# Importar la librer칤a
import statsmodels.api as sm

# OJO hay que a침adir el vector de unos
# Hay un m칠todo espec칤fico en stasmodels para a침adir el vector de unos
X = sm.add_constant(x)

# Instanciar y ajustar el modelo
modelo = sm.OLS(y,X).fit()

# Imprimir los resultados
print(modelo.summary())
```

```python
# Obtener unicamente las betas
modelo.params
```

```python
# Obtener sigma (no hay un m칠todo directo)
# Obtener sse=sum squared error y luego calcular sigma

sigma_sm=np.sqrt(modelo.ssr/(observaciones-2))
sigma_sm
```

```python
# Juntar las betas y sigma en un array

parametros_sm=np.append([modelo.params], [sigma_sm])
parametros_sm
```

# Scikit-Learn

Tambi칠n podemos calcular los par치metros de la regresi칩n lineal con la librer칤a `sklearn`



```python
# Importar la librer칤a
from sklearn import linear_model

# Preparar los datos (convertir x en una matriz nx1)
x=x.reshape(-1, 1)

# Instanciar el modelo
modelo = linear_model.LinearRegression()

# Ajustar el modelo
modelo.fit(x,y)

# Obtener los par치metros
print('beta0: ',modelo.intercept_,'\nbeta1: ', modelo.coef_)
```

```python
# Sklearn no calcula sigma. Lo calculamos con la f칩rmula
sigma_sk=np.sqrt(sum((y-modelo.predict(x))**2)/(observaciones-2))
sigma_sk
```

```python
# Crear un array con los tres par치metros. np.append admite solo dos arrays
# Para unir tres arrays, unimos dos y luego otro al resultado

parametros_sk=np.append(np.append([modelo.intercept_],[modelo.coef_]),[sigma_sk])
parametros_sk
```

# Resumen de los par치metros

Los par치metros iniciales son los par치metros "verdaderos", $$(\beta_0,\beta_1,\sigma)$$

En un problema real, los par치metros "verdaderos" se supone que existen, pero no son observables.

Los par치metros que se obtienen son estimaciones de los pr치metros "verdaderos".

$$(\hat{\beta_0},\hat{\beta_1},\hat{\sigma}) $$
